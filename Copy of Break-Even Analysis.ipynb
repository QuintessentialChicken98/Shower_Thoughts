```python
# import modules
import matplotlib.pyplot as  plt
import matplotlib as mpl
import matplotlib.font_manager as font_manager
import numpy as np
import seaborn as sns

# Set font to LaTeX 'Computer Modern': 
mpl.rcParams['font.family']='serif'
cmfont = font_manager.FontProperties(fname=mpl.get_data_path() + '/fonts/ttf/cmr10.ttf')
mpl.rcParams['font.serif']=cmfont.get_name()
mpl.rcParams['mathtext.fontset']='cm'
mpl.rcParams['axes.unicode_minus']=False
```

```python
# LaTeX Style:
# Set parameters:
q = np.arange(0,100)
s_tc, i_tc = 2, 20 # slope & intercept (line 1)
s_tr, i_tr = 2.5, 0 # slope & intercept (line 2)
tc = (q*s_tc+i_tc)
tr = (q*s_tr+i_tr)

# Build plot:
plt.figure(figsize=(10,5), dpi=150)
plt.xlim(0, q.max())
plt.ylim(0, tr.max())
# plt.axhline(20, label = 'Fixed Costs', color = 'black', linestyle=':')
plt.plot(tr, label = 'Total Revenue', color='black')
plt.plot(tc, label='Total Costs', color='black', linestyle='--')

# Intersection point:
xi = (i_tc-i_tr) / (s_tr-s_tc)
yi = s_tc * xi + i_tc
plt.scatter(xi,yi, color='black', label = 'Break-Even Point')
plt.hlines(yi, xmin=0, xmax=xi,  color = 'black', linewidth=1, linestyle=':')
plt.vlines(xi, ymin=0, ymax=yi, color = 'black', linewidth=1, linestyle=':')

# Fill profit and loss:
loss = plt.fill_between(q, tc, tr,where = tr<tc, color='grey', alpha=.5)
(x0, y0), (x1, y1) = loss.get_paths()[0].get_extents().get_points()
plt.text((x0 + x1) / 2, (y0 + y1) / 2, "Loss", ha='right', va='center', rotation = 25, fontsize=10, fontweight='bold')
profit = plt.fill_between(q, tc, tr, where = tr>tc, color='grey', alpha=.5)
(x2, y2), (x3, y3) = profit.get_paths()[0].get_extents().get_points()
plt.text((x2 + x3) / 2, (((y2 + y3)/2)-21), "Profit", ha='right', va='bottom', rotation = 25, fontsize=10, fontweight='bold')
# plt.text(80, 195, "Profit", size=10, rotation=25, ha="right", va="top")

# Plot labels:
plt.ylabel('P', rotation = 0)
plt.gca().yaxis.set_label_coords(-0.05,1.02)
plt.xlabel('Q')
plt.gca().xaxis.set_label_coords(1.02,-0.05)
plt.title('Break-Even Analysis', size=14,fontweight="bold")
plt.legend()
plt.show();
```
